```python
from graphviz import Digraph
import pdb
```

```python
dot = Digraph(comment='Once Upon a Time')
```

```python
def add_node(graph, nodes):
    for n in range(0, len(nodes)):
        if len(nodes[n]) == 1:
            graph.node(nodes[n][0])  
        elif len(nodes[n]) == 2:
            graph.node(nodes[n][0], nodes[n][1])
        else:
            quit('incorrect number of arguments for names of nodes')
    return graph   
```

```python
def add_edge(graph, edges):
    for n in range(0, len(edges)): 
        if len(edges[n]) == 2:
            graph.edge(edges[n][0], edges[n][1])
        elif len(edges[n]) == 3:
            graph.edge(edges[n][0], edges[n][1], edges[n][2])
        elif len(edges[n]) == 4:
            graph.edge(edges[n][0], edges[n][1], edges[n][2], constraint=edges[n][3]) 
            
        else:
            quit('incorrect number of arguments for edges')
    return graph
```
